```python
# Step 1: Install Libraries (if not already installed)
!pip install pandas matplotlib seaborn --quiet

# Step 2: Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 3: Create Synthetic Dataset
data = {
    "Food": ["Apple","Banana","Orange","Broccoli","Carrot","Spinach","Chicken Breast",
             "Salmon","Egg","Almonds","Walnuts","Brown Rice","Oats","Cheddar Cheese",
             "Yogurt","Bread","Butter","Potato","Tomato","Peanut Butter"],
    "Category": ["Fruit","Fruit","Fruit","Vegetable","Vegetable","Vegetable","Meat",
                 "Fish","Protein","Nuts","Nuts","Grain","Grain","Dairy","Dairy",
                 "Grain","Dairy","Vegetable","Vegetable","Nuts"],
    "Protein":[0.3,1.3,0.9,2.8,0.9,2.9,31,20,13,21,15,2.6,17,25,10,9,0.9,2,0.9,25],
    "Fat":[0.2,0.3,0.1,0.4,0.2,0.4,3.6,13,11,49,65,0.9,7,33,0.4,3.2,81,0.1,0.2,50],
    "Carbs":[14,27,12,7,10,3.6,0,0,1.1,22,14,23,66,1.3,4.7,49,0.1,17,3.9,20],
    "Sugar":[10,14,9,1.5,5,0.4,0,0,1.1,4.4,2.6,0.4,1,0.5,4.7,5,0.1,0.8,2.6,9],
    "Calories":[52,105,47,34,41,23,165,208,155,576,654,112,389,403,59,265,717,77,18,588]
}

df = pd.DataFrame(data)

# Step 4: Quick Look at Data
print(df.head())
print(df.info())

# Step 5: Exploratory Data Analysis (EDA)

# 1. Top 10 Foods by Protein
top_protein = df[['Food','Protein']].sort_values(by='Protein', ascending=False).head(10)
top_protein.plot.bar(x='Food', y='Protein', title='Top 10 Foods by Protein', figsize=(10,6))
plt.show()

# 2. Top 10 Foods by Fat
top_fat = df[['Food','Fat']].sort_values(by='Fat', ascending=False).head(10)
top_fat.plot.bar(x='Food', y='Fat', title='Top 10 Foods by Fat', figsize=(10,6), color='orange')
plt.show()

# 3. Top 10 Foods by Carbs
top_carbs = df[['Food','Carbs']].sort_values(by='Carbs', ascending=False).head(10)
top_carbs.plot.bar(x='Food', y='Carbs', title='Top 10 Foods by Carbohydrates', figsize=(10,6), color='green')
plt.show()

# 4. Distribution of Macronutrients
sns.histplot(df['Protein'], kde=True, color='blue', label='Protein')
sns.histplot(df['Fat'], kde=True, color='orange', label='Fat')
sns.histplot(df['Carbs'], kde=True, color='green', label='Carbs')
plt.legend()
plt.title("Distribution of Macronutrients")
plt.show()

# 5. Correlation Between Macronutrients
plt.figure(figsize=(8,6))
sns.heatmap(df[['Protein','Fat','Carbs']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Macronutrients")
plt.show()

# Step 6: Save Dataset for GitHub or Later Use
df.to_csv("food_nutrition_dataset.csv", index=False)
print("we got the result successfully.")
```
